Ara afegirem una nova columna que es digui H2H, que serà cuants cops A ha guanyat a B, a la inversa i la diferencia.

```python
df_final = pd.read_excel('all_data_with_odds.xlsx')

#Filtem fins abans del torneig
df_final['Date'] = pd.to_datetime(df_final['Date'], format='%A, %B %d, %Y', errors='coerce')
df_final = df_final[df_final['Date'] < data_tall]

df_final.head()
```

```python
import pandas as pd

# 1. Copia
df = df_final.copy()

# 2. WinnerBinary a int
df['WinnerBinary'] = df['WinnerBinary'].astype(int)

# 3. Crear parell ordenat (A < B alfabeticamente)
df['Pair'] = df.apply(
    lambda row: tuple(sorted([row['PlayerA'], row['PlayerB']])), 
    axis=1
)

# 4. Guayador real
df['Ganador'] = df.apply(
    lambda row: row['PlayerA'] if row['WinnerBinary'] == 1 else row['PlayerB'],
    axis=1
)

# 5. Contar H2H
h2h = df.groupby(['Pair', 'Ganador']).size().unstack(fill_value=0)

# --- CLAVE: Asegurar totes las columnas ---
all_players = pd.Index(df['PlayerA'].unique()).union(df['PlayerB'].unique())
h2h = h2h.reindex(columns=all_players, fill_value=0)

# 6. Funció CORREGIDA: fer servir .at[] (escalar)
def get_h2h(row):
    pair = row['Pair']
    a = row['PlayerA']
    b = row['PlayerB']
    
    if pair in h2h.index:
        a_wins = h2h.at[pair, a]   # ← ESCALAR
        b_wins = h2h.at[pair, b]   # ← ESCALAR
    else:
        a_wins = b_wins = 0
    
    return pd.Series([a_wins, b_wins, a_wins - b_wins])

# 7. Aplicar
df[['H2H_A_wins', 'H2H_B_wins', 'H2H_Diff']] = df.apply(get_h2h, axis=1)

# 8. Netejar
df.drop(['Pair', 'Ganador'], axis=1, inplace=True)

# 9. Veure resultats
df[['PlayerA', 'PlayerB', 'Winner', 'H2H_A_wins', 'H2H_B_wins', 'H2H_Diff']].head(10)
```

```python
df_h2h = df.copy()

df_h2h = df_h2h[df_h2h['Tournament'] == 'Wimbledon']

df_h2h = df_h2h[['PlayerA', 'PlayerB', 'RankA', 'RankB', 'WinnerBinary', 'MaxOddsPlayerA', 'MaxOddsPlayerB', 'AvgOddsPlayerA', 'AvgOddsPlayerB', 'H2H_A_wins', 'H2H_B_wins', 'H2H_Diff']]
df_h2h.head()
```

```python
df_h2h.to_csv('wimbeldon_bets_h2h.csv', index=False)
```
